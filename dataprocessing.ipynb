```python
import pandas as pd
data=pd.read_csv('Book1.csv - Sheet1.csv')
data.head()
```

Handling Missing Values

```python
data.fillna({
    'Age':data['Age'].mean(),
    'marks':data['marks'].mean(),
    'city':'Unknown'
},inplace=True)
print(data)
```
